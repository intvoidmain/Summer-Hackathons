```python
"""
Task 3

@author: alias890
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv(r'C:\Users\Varu\Desktop\House_prediction.csv')

x= df.to_numpy()

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_x = LabelEncoder()
x[:, 0] = labelencoder_x.fit_transform(x[:,0])
x[:, 6] = labelencoder_x.fit_transform(x[:,6])
x[:, 7] = labelencoder_x.fit_transform(x[:,7])

y = x[:,10]
X= x[:,0:5]


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

X_train,X_test, y_train, y_test = train_test_split(X, y, test_size = 10000/10691, random_state = 0)


regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

data = pd.DataFrame({'Actual':y_test.flatten(), 'Predicted':y_pred.flatten()})



data1 = data.sample(n = 100)
data1.plot(kind='bar',figsize =(10,8))
plt.grid(which = 'major',linestyle = '-', color = 'green')
plt.grid(which = 'minor', linestyle = ':',linewidth = '0.5', color = 'black')
plt.show()


```
